```python
import pandas as pd
import matplotlib.pyplot as plt


x_df = pd.read_csv("accx.csv") 
y_df = pd.read_csv("accy.csv")
z_df = pd.read_csv("accz.csv")

x_series = x_df["value"].rename("accelerometer_x")
y_series = y_df["value"].rename("accelerometer_y")
z_series = z_df["value"].rename("accelerometer_z")


n = min(len(x_series), len(y_series), len(z_series))
x_series = x_series.iloc[:n].reset_index(drop=True)
y_series = y_series.iloc[:n].reset_index(drop=True)
z_series = z_series.iloc[:n].reset_index(drop=True)



# Plot X
plt.figure(figsize=(10,4))
plt.plot(x_series, label="Accelerometer X")
plt.xlabel("Sample Index")
plt.ylabel("X-axis Value")
plt.title("Accelerometer X")
plt.legend()
plt.show()

# Plot Y
plt.figure(figsize=(10,4))
plt.plot(y_series, label="Accelerometer Y", color="orange")
plt.xlabel("Sample Index")
plt.ylabel("Y-axis Value")
plt.title("Accelerometer Y")
plt.legend()
plt.show()

# Plot Z
plt.figure(figsize=(10,4))
plt.plot(z_series, label="Accelerometer Z", color="green")
plt.xlabel("Sample Index")
plt.ylabel("Z-axis Value")
plt.title("Accelerometer Z")
plt.legend()
plt.show()

# combined plot
plt.figure(figsize=(10,5))
plt.plot(x_series, label="X-axis")
plt.plot(y_series, label="Y-axis")
plt.plot(z_series, label="Z-axis")
plt.xlabel("Sample Index")
plt.ylabel("Accelerometer Values")
plt.title("Accelerometer X, Y, Z Combined")
plt.legend()
plt.show()
```
