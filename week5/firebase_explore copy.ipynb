```python
import time, random
import firebase_admin
from firebase_admin import credentials, db

KEY_PATH = "/Users/arsalansharifizad/Documents/code/sit225-project-906a4-firebase-adminsdk-fbsvc-aa4e493a52.json"
DB_URL   = "https://sit225-project-906a4-default-rtdb.asia-southeast1.firebasedatabase.app/"

SENSOR_ID = "DHT22-1234"


if not firebase_admin._apps:
    cred = credentials.Certificate(KEY_PATH)
    firebase_admin.initialize_app(cred, {"databaseURL": DB_URL})


db.reference(f"/sensors/{SENSOR_ID}/meta").set({
    "sensor_name": SENSOR_ID,
    "type": "DHT22",
    "created_at_ms": int(time.time() * 1000)
})

def add_dht22_reading(sensor_id: str, temp_c: float, hum_pct: float):
    payload = {
        "sensor_name": sensor_id,
        "timestamp_ms": int(time.time() * 1000),
        "temperature_c": float(temp_c),
        "humidity_pct": float(hum_pct),
    }
    db.reference(f"/sensors/{sensor_id}/readings").push(payload)
    return payload

N = 10
for _ in range(N):
    temp = random.uniform(10.0, 40.0) 
    hum  = random.uniform(10.0, 55.0)
    inserted = add_dht22_reading(SENSOR_ID, temp, hum)
    print("Inserted:", inserted)
    time.sleep(0.5)  
```

```python
import time, random
import firebase_admin
from firebase_admin import credentials, db

KEY_PATH = "/Users/arsalansharifizad/Documents/code/sit225-project-906a4-firebase-adminsdk-fbsvc-aa4e493a52.json"
DB_URL   = "https://sit225-project-906a4-default-rtdb.asia-southeast1.firebasedatabase.app/"

SENSOR_ID = "SR04-0001"


if not firebase_admin._apps:
    cred = credentials.Certificate(KEY_PATH)
    firebase_admin.initialize_app(cred, {"databaseURL": DB_URL})


db.reference(f"/sensors/{SENSOR_ID}/meta").set({
    "sensor_name": SENSOR_ID,
    "type": "sr04",
    "created_at_ms": int(time.time() * 1000)
})

def add_sr04_reading(sensor_id: str, distance_cm: float):
    payload = {
        "sensor_name": sensor_id,
        "timestamp_ms": int(time.time() * 1000),
        "distance_cm": float(distance_cm)
    }
    db.reference(f"/sensors/{sensor_id}/readings").push(payload)
    return payload


N = 12
baseline = 60.0
for _ in range(N):

    baseline += random.uniform(-4.0, 4.0)
    distance = max(40.0, min(120.0, baseline))
    inserted = add_sr04_reading(SENSOR_ID, distance)
    print("Inserted:", inserted)
    time.sleep(0.5)  
```
