```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv("gyro_data.csv", encoding="utf-8-sig")
df["gx_dps"] = pd.to_numeric(df["gx_dps"], errors="coerce")

if "timestamp_iso" in df.columns:
    df["timestamp_iso"] = pd.to_datetime(df["timestamp_iso"], errors="coerce")
    x = df["timestamp_iso"]
    x_label = "Time"
else:
    x = df.index
    x_label = "Sample Index"

plt.figure(figsize=(10,5))
plt.plot(x, df["gx_dps"], label="Gyro X (gx_dps)", color="red")
plt.title("Gyroscope X over Time")
plt.xlabel(x_label)
plt.ylabel("Degrees per second (dps)")
plt.legend()
if "timestamp_iso" in df.columns:
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%d %H:%M:%S"))
    plt.gcf().autofmt_xdate()
plt.show()
```

```python
df["gy_dps"] = pd.to_numeric(df["gy_dps"], errors="coerce")

if "timestamp_iso" in df.columns:
    x = df["timestamp_iso"]
    x_label = "Time"
else:
    x = df.index
    x_label = "Sample Index"

plt.figure(figsize=(10,5))
plt.plot(x, df["gy_dps"], label="Gyro Y (gy_dps)", color="green")
plt.title("Gyroscope Y over Time")
plt.xlabel(x_label)
plt.ylabel("Degrees per second (dps)")
plt.legend()
if "timestamp_iso" in df.columns:
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%d %H:%M:%S"))
    plt.gcf().autofmt_xdate()
plt.show()
```

```python
df["gz_dps"] = pd.to_numeric(df["gz_dps"], errors="coerce")

if "timestamp_iso" in df.columns:
    x = df["timestamp_iso"]
    x_label = "Time"
else:
    x = df.index
    x_label = "Sample Index"

plt.figure(figsize=(10,5))
plt.plot(x, df["gz_dps"], label="Gyro Z (gz_dps)", color="blue")
plt.title("Gyroscope Z over Time")
plt.xlabel(x_label)
plt.ylabel("Degrees per second (dps)")
plt.legend()
if "timestamp_iso" in df.columns:
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%d %H:%M:%S"))
    plt.gcf().autofmt_xdate()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(x, df["gx_dps"], label="X (gx_dps)", color="red")
plt.plot(x, df["gy_dps"], label="Y (gy_dps)", color="green")
plt.plot(x, df["gz_dps"], label="Z (gz_dps)", color="blue")
plt.title("Gyroscope Data (X, Y, Z)")
plt.xlabel(x_label)
plt.ylabel("Degrees per second (dps)")
plt.legend()
if "timestamp_iso" in df.columns:
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%d %H:%M:%S"))
    plt.gcf().autofmt_xdate()
plt.show()
```
